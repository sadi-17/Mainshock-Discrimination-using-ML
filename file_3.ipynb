```python
import numpy as np
import pandas as pd
from pathlib import Path
from scipy.spatial import cKDTree

REF_LABEL = "label_R50_T30"
CATS = ["comcat_global_m4", "scedc_socal_m25", "comcat_himalaya_m4"]
MODELS_DIR = Path("./models")
FEAT_R_KM = 50.0
EARTH_R = 6371.0
N_LOCAL = 50            # events in the local b-value window
MIN_EVENTS = 15         # need at least this many to estimate b
BACKGROUND_N = 500      # events for the rolling background b-value


# ---- DeLong (inlined, validated in step 11) ----
def _midrank(x):
    order = np.argsort(x); z = x[order]; n = len(x); t = np.zeros(n); i = 0
    while i < n:
        j = i
        while j < n and z[j] == z[i]:
            j += 1
        t[i:j] = 0.5 * (i + j - 1) + 1; i = j
    out = np.empty(n); out[order] = t; return out

def delong_test(y, p1, p2):
    from scipy.stats import norm
    m, n = int((y == 1).sum()), int((y == 0).sum())
    if m == 0 or n == 0:
        return np.nan, np.nan, np.nan
    preds = np.vstack([p1, p2]); tx = np.empty((2, m)); ty = np.empty((2, n))
    tz = np.empty((2, m + n))
    for r in range(2):
        px, py = preds[r][y == 1], preds[r][y == 0]
        tz[r] = _midrank(np.concatenate([px, py])); tx[r] = _midrank(px)
        ty[r] = _midrank(py)
    aucs = tz[:, :m].sum(axis=1) / (m * n) - (m + 1.0) / (2.0 * n)
    v01 = (tz[:, :m] - tx) / n; v10 = 1.0 - (tz[:, m:] - ty) / m
    S = np.cov(v01) / m + np.cov(v10) / n
    d = np.array([1.0, -1.0]); var = d @ S @ d
    if var <= 0:
        return float(aucs[0]), float(aucs[1]), 1.0
    z = (aucs[0] - aucs[1]) / np.sqrt(var)
    return float(aucs[0]), float(aucs[1]), float(2 * norm.sf(abs(z)))


def aki_b(mags, mc):
    """Aki (1965) maximum-likelihood b-value; needs mean(m) > mc."""
    m = mags[mags >= mc - 1e-9]
    if len(m) < MIN_EVENTS:
        return np.nan
    mean_m = m.mean()
    if mean_m - mc <= 1e-6:
        return np.nan
    return 1.0 / (np.log(10) * (mean_m - mc))


def to_xyz(lat, lon):
    lat, lon = np.radians(lat), np.radians(lon)
    return np.column_stack([np.cos(lat)*np.cos(lon),
                            np.cos(lat)*np.sin(lon), np.sin(lat)])

def chord(dkm): return 2*np.sin(dkm/(2*EARTH_R))


def local_bvalues(df):
    """For each event, b-value of the preceding N_LOCAL local events, and a
    rolling background b over the preceding BACKGROUND_N events (any location)."""
    n = len(df)
    times = ((df["time"] - pd.Timestamp(0, tz="UTC")).dt.total_seconds()
             .to_numpy())
    mw = df["mw"].to_numpy(float)
    xyz = to_xyz(df["latitude"].to_numpy(), df["longitude"].to_numpy())
    tree = cKDTree(xyz)
    r = chord(FEAT_R_KM)
    local_b = np.full(n, np.nan)
    for i in range(n):
        nb = np.asarray(tree.query_ball_point(xyz[i], r))
        nb = nb[(nb < i)]                      # strictly past
        if len(nb) < MIN_EVENTS:
            continue
        nb = nb[np.argsort(times[nb])[-N_LOCAL:]]
        mags = mw[nb]
        local_b[i] = aki_b(mags, mags.min())
    # rolling background b over the last BACKGROUND_N events by time
    bg_b = np.full(n, np.nan)
    for i in range(n):
        lo = max(0, i - BACKGROUND_N)
        mags = mw[lo:i]
        if len(mags) >= MIN_EVENTS:
            bg_b[i] = aki_b(mags, mags.min())
    return local_b, bg_b


def main():
    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import roc_auc_score, f1_score
    rows = []
    for cat in CATS:
        spl = np.load(MODELS_DIR / f"{cat}__splits.npz")
        b1, b3 = int(spl["b1"]), int(spl["b3"])
        df = pd.read_parquet(f"{cat}_features.parquet").sort_values("time")\
               .reset_index(drop=True)
        y = df[REF_LABEL].to_numpy(np.int8)

        cache = MODELS_DIR / f"{cat}__bvalues.npz"
        if cache.exists():
            d = np.load(cache); local_b, bg_b = d["local_b"], d["bg_b"]
        else:
            print(f"computing b-values for {cat} ...")
            local_b, bg_b = local_bvalues(df)
            np.savez(cache, local_b=local_b, bg_b=bg_b)

        # b-value drop score (Gulia-Wiemer rationale)
        drop = bg_b - local_b
        model_p = np.load(MODELS_DIR / f"{cat}__xgboost__recal_probs.npz")\
                    ["prob"].astype(float)

        # evaluate on test events where b-value is defined (fair subset)
        te = np.arange(b3, len(df))
        valid = te[np.isfinite(drop[te])]
        cov = len(valid) / len(te)
        yv = y[valid]

        # 1-feature logistic on local_b, fit on train subset where defined
        tr = np.arange(0, b1)
        tr = tr[np.isfinite(local_b[tr])]
        lr = LogisticRegression(max_iter=1000)
        lr.fit(local_b[tr].reshape(-1, 1), y[tr])
        p_lr = lr.predict_proba(local_b[valid].reshape(-1, 1))[:, 1]

        for name, score in [("b-drop rule", drop[valid]),
                            ("b-value logistic", p_lr)]:
            a_m, a_b, p = delong_test(yv, model_p[valid], score)
            rows.append({"catalog": cat, "baseline": name,
                         "test_coverage": round(cov, 3),
                         "auc_baseline": round(a_b, 4),
                         "auc_model_same_subset": round(a_m, 4),
                         "delong_p": p})
            print(f"  {cat} | {name}: AUC {a_b:.3f} "
                  f"(model {a_m:.3f} on same subset, p={p:.2e}, "
                  f"coverage {cov:.1%})")

    out = pd.DataFrame(rows)
    out.to_csv("step13_bvalue_baseline.csv", index=False)
    print("\n=== step13_bvalue_baseline.csv ===")
    print(out.to_string(index=False))


if __name__ == "__main__":
    main()
```
